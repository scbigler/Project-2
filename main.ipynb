```python
import pandas as pd
```

```python
df1 = pd.read_csv('Resources/data-2.csv')
```

```python
df2 = pd.read_csv('Revenue.csv')
df2['Manufacturer'] = df2['Manufacturer'].str.upper()
df2
```

```python
df3 = df1.filter(items=['covered_recipient_first_name','covered_recipient_last_name','date_of_payment','applicable_manufacturer_or_applicable_gpo_making_payment_name','total_amount_of_payment_usdollars','number_of_payments_included_in_total_amount'])
df3
```

```python
company_df = df1.filter(items=['applicable_manufacturer_or_applicable_gpo_making_payment_name','total_amount_of_payment_usdollars']).groupby(df3['applicable_manufacturer_or_applicable_gpo_making_payment_name'].astype(str).str.upper())
company_df2 = company_df['total_amount_of_payment_usdollars'].sum().reset_index().rename(columns={'applicable_manufacturer_or_applicable_gpo_making_payment_name':'Manufacturers', 'total_amount_of_payment_usdollars':'total_payment_in_2019'})
company_df2
```

```python
df2['join'] = 1
company_df2['join'] = 1

combine_df = df2.merge(company_df2, on='join').drop('join', axis=1)
company_df2.drop('join', axis=1, inplace=True)

combine_df['match'] = combine_df.apply(lambda x: x.Manufacturers.find(x.Manufacturer), axis=1).ge(0)
```

```python
final_df = combine_df.loc[combine_df['match'] == True].reset_index(drop=True)

final_df
```

```python
final_df.drop(['Manufacturers', 'match'], axis = 1, inplace = True)
final_df
```
